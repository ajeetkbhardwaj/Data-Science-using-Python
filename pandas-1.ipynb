# 2.0 Creating Pandas Series

```python
# create a simple pandas series from a list
a = [1 ,3 ,4]
myser = pd.Series(a)
print(myser)

# Creating label and accessing particular value from series
print(myser[2])

myser1 = pd.Series(a,index = ["x","y","z"])
print(myser1)

print(myser1["y"])
```

```python
calories = {"day1": 420, "day2": 180, "day3":390}
myenergy = pd.Series(calories)
print(myenergy)

myenergy1 = pd.Series(calories,index = ["day1","day2"])
print(myenergy1)
```

# 3.0 Pandas DataFrames

```python
# to set  DataFrame from two Series
data = {
    "calories" : [420 , 380 , 390],
    "duration" : [50 , 40 , 45]
}
df = pd.DataFrame(data)
print(df)
```

```python
# reference to the row index :
print(df.loc[0])

# use a list of index :
print(df.loc[[0, 1]])

# name the index :
df = pd.DataFrame(data, index = ["day1","day2","day3"])
print(df)
```
